```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_digits
from sklearn.naive_bayes import ComplementNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import CategoricalNB
```

```python
digits=load_digits()
```

```python
g=RandomForestClassifier()
g.get_params()
```

```python
model_para = {'LogisticRegression':
              {'model':LogisticRegression(solver='lbfgs',max_iter=200),'para':
               {
                'C':[100, 10, 1.0, 0.1, 0.01]
                }
               },
              'GaussianNB':
              {
              'model':GaussianNB(),'para':
              {
                  
               }
               },
              'MultinomialNB':{
                  'model':MultinomialNB(),'para':
                  {
                      'alpha':[1.0,5.0,9.0]
                  }
              },
              'svm':{'model':svm.SVR(),'para':
                     {
                         'C':[1.0,2.0,3.0,4.0]
                        
                     }
                     },
              'DecisionTree':{'model':DecisionTreeClassifier(),'para':
                              {'ccp_alpha':[0.0,0.5,1.0]
                             
                                  
                              }
                              },
              'RandomForest':{'model':RandomForestClassifier(),'para':
                              {
                                'n_estimators':[5,10,50,70],
                                 
                              }
                              },
              'BernoulliNB':{'model':BernoulliNB(),'para':
                             {
                                 'alpha':[0,0.5,1,1.5,2],
                              
                             }
                             },
              'ComplementNB':{'model':ComplementNB(),'para':
                              {
                                  'alpha':[0,0.5,1,1.5,2],
                               'norm':[True,False]
                              }
                              },
              'CatergoricalNB':{'model':CategoricalNB(),'para':
                               {
                                   'alpha':[0,0.5,1,1.5,2],
                                'min_categories':[None,3,4,5]
                               }
                               }
              

}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
score=[]
for model_name,mp in model_para.items():
  clf = GridSearchCV(mp['model'],mp['para'],cv=5,return_train_score=False)
  clf.fit(digits.data,digits.target)
  score.append({'model':model_name,
                'best score':clf.best_score_,
                'best para':clf.best_params_
                })
score
```

```python
df=pd.DataFrame(score)
df
```

```python
param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000] }
c=GridSearchCV(LogisticRegression(solver='lbfgs',max_iter=900),param_grid,cv=5,return_train_score=False)
c.fit(digits.data,digits.target)
c.best_score_
```
